```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.patches import Circle
from matplotlib.collections import PatchCollection
```

```python
r_lsst = 1.75
r_circ = 4.0
theta = np.radians(np.arange(0, 360, 0.5))
```

```python
fig = plt.figure(figsize=(8, 8))
plt.axis('equal')
ax = plt.subplot(1,1,1)
x = r_circ * np.sin(theta)
y = r_circ * np.cos(theta)
#plt.plot(x, y, 'b-', linewidth=2)
p = PatchCollection([Circle((0, 0), radius=r_circ, fc='b', ec='b')], alpha=0.4)
ax.add_collection(p)
x = r_lsst * np.sin(theta)
y = r_lsst * np.cos(theta)
plt.plot(x, y, 'r-')
r2 = 3.0
theta2 = np.radians(np.arange(0, 360, 360/6))
for t in theta2:
    offset_x = r2 * np.sin(t)
    offset_y = r2 * np.cos(t)
    x = r_lsst * np.sin(theta) + offset_x
    y = r_lsst * np.cos(theta) + offset_y
    plt.plot(x, y, 'r-')
plt.xlabel(r'Degrees', fontsize='x-large')
plt.ylabel(r'Degrees', fontsize='x-large')
plt.title(r'LSST tiling of ToO location', fontsize='xx-large')
plt.savefig('ToO_tiling.pdf')
```
